```python
import pandas as pd

# caminho do arquivo
file_path = '../data/raw/dados_2025-02-05.xlsx'

#df = pd.read_csv(file_path)
df = pd.read_excel(file_path)

df.head()
```

```python
df.info()

df.describe()

df.isnull().sum()
```

```python
#Transform

# remover duplicados
df = df.drop_duplicates()

# remover nulos
df = df.dropna()

# converter datas
df['data'] = pd.to_datetime(df['data'], errors='coerce')

# padronizar texto
df['nome'] = df['nome'].str.strip().str.upper()
```

```python
# Save Load Final

output_path = '../data/processed/dados_tratados.csv'

df.to_csv(output_path, index=False)
```
