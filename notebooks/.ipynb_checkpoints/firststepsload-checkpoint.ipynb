```python
import pandas as pd
import os

from pathlib import Path

arquivo = Path("../data/raw/CargosVagosVacancias_202603.ods")

print(arquivo.exists())


# Ver quais abas existem
xls = pd.ExcelFile(arquivo, engine="odf")

print(xls.sheet_names)
```

```python
# ler o ODS ( Open Document SUSSHETS)
df = pd.read_excel(
    arquivo,
    engine="odf"
)

print(df.head())
print("========")
df.info()
print("========")
df.describe()
print("========")
df.isnull().sum()
```

```python
# Ler uma aba especifica 
# aba 1 - por_órgão
df_orgao = pd.read_excel(
    arquivo,
    sheet_name='por_Órgão',
    engine='odf'
)

print(df_orgao.head())
```

```python
# Ler uma aba especifica 
# aba 2 - por_órgão_e_Cargo
df_cargo = pd.read_excel(
    arquivo,
    sheet_name='por_Órgao_e_Cargo',
    engine='odf'
)

print(df_cargo.head())
```

```python
# ver tamanho

print(df_orgao.shape)
print(df_cargo.shape)
print("=====")
print(df_cargo.columns)
print("===")
print(df_cargo.info())
```
